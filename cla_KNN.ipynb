```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
root="/mnt/MyDrive/AI"
path_dataset=root+"/dataset/"
dataset_file="Titanic-Dataset.csv"
```

```python
df=pd.read_csv(path_dataset+dataset_file)
```

```python
df.head(10)
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#Hapus kolom kabin karena terlalu banyak yg kosong(>50%)
df.drop(columns=['Cabin'],inplace=True)
```

```python
df.info()
```

```python
mean_age=df.Age.mean()
mean_age
```

```python
df.Age.fillna(mean_age,inplace=True)
```

```python
df.Embarked.value_counts()
```

```python
df.Embarked.fillna('S',inplace=True)
```

```python
#Cek duplikasi data
df.drop_duplicates()
```

```python
df.Pclass.value_counts()
```

```python
col_cate=['Pclass','Sex','Embarked']
```

```python
fig,axes=plt.subplots(1,3, figsize=(12,5))
axes=axes.flatten()
for cate,axs in zip(col_cate,axes):
  axs.set_title(cate,fontsize=14)
  sn.countplot(x=cate,data=df, ax=axs,hue='Survived')
```
